```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (replace this with your actual data)
model_names = ['EDVR', 'BasicVSR', 'IconVSR']
psnr_scores = [30.4261 , 31.4255, 31.7017 ]
ssim_scores = [ 0.8690 ,0.8915,0.8957 ]

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot PSNR on the first y-axis
color = 'tab:red'
ax1.set_xlabel('Models')
ax1.set_ylabel('PSNR', color=color)
ax1.plot(model_names, psnr_scores, color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for SSIM
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('SSIM', color=color)
ax2.plot(model_names, ssim_scores, color=color, marker='s')
ax2.tick_params(axis='y', labelcolor=color)

# Show the plot
plt.title('Comparison of Models by PSNR & SSIM ')
plt.show()

```
